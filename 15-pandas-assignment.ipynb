```python
#creating  dataframe whose index the first 10 letters of the alphabet
import string
alphabet = list(string.ascii_lowercase)
alphabet10 = alphabet[:10]
print(alphabet10)

```

```python
#first 10 prime numbers.
prime = []
for num in range(2,101):
    if all(num%i!=0 for i in range(2,num)):
        prime.append(num)
prime10 = prime[:10]
print(prime10)
```

```python
#first 10 fibonacci numbers
a,b = 0,1
fib10 = []
for i in range(20):
    fib10.append(a)
    a,b = b,a+b
#print(fib10)
fib10 = [i for n, i in enumerate(fib10) if i not in fib10[:n]]
fib10 = fib10[:10]
print(fib10[:10])
```

```python
#creating a dataframe whose index is the first 10 letters of the alphabet 
#and columns are the first 10 prime numbers and first 10 fibonacci numbers
new_df = {
    'prime numbers': prime10,
    'fibonacci numbers': fib10
}
new_df = pd.DataFrame(new_df, index=alphabet10)
print(new_df)
```

```python
flights = pd.read_csv('flights.csv')
airline = pd.read_csv('airlines.csv')
flights.head()
```

```python
airline.head()
```

```python
#Merge these two DataFrames into a single DataFrame: nycflights13.
nycflights13 = pd.merge(flights, airline, on='carrier', how='inner')
nycflights13 = pd.merge(flights, airline, on='carrier', how='outer')
nycflights13.head()
```

```python
# other way to merge two dataframes
nycflights13 = flights.set_index('carrier').join(airline.set_index('carrier'))
nycflights13.head()
```
